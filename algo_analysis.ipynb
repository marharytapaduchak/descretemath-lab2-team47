Аналіз Часу Виконанн
1. DFS: Порівняння ітеративного та рекурсивного підходів:
- Ітеративний DFS зі словником суміжності
    - Найшвидший метод для DFS серед усіх, особливо на малих графах (200 вузлів).
- Ітеративний DFS з матрицею суміжності:
    - Час виконання для 1000 вузлів складає 0.1 секунди, що в 5 разів повільніше за ітеративний DFS зі словником (0.02 секунди).
- Рекурсивний DFS:
    - Обидва підходи (зі словником і матрицею) повільніші за ітеративні.
2. BFS: Ітеративний підхід
- Ітеративний BFS зі словником суміжності та  з матрицею суміжності:
    - Як і в DFS функції з словниками швидші ніж функції з матрицями(лісти).
    - Можна побачити що BFS є швидший на обох видах даних в цій ситуації


```python
import matplotlib.pyplot as plt

# Node sizes
node_sizes = [200, 500, 1000]

# DFS Data
dfs_iter_dict = [0.0009, 0.005, 0.02]
dfs_iter_matrix = [0.004, 0.03, 0.1]
dfs_rec_dict = [0.01, 0.15, 1.2]
dfs_rec_matrix = [0.01, 0.2, 1.27]

# Corrected BFS Data
bfs_iter_dict = [0.0003, 0.001, 0.004]
bfs_iter_matrix = [0.001, 0.01, 0.05]

# Plot 1: Overall DFS vs BFS (Separate Comparison)
plt.figure(figsize=(12, 6))

# DFS data
plt.plot(node_sizes, dfs_iter_dict, marker='o', label='DFS Iterative (Dict)')
plt.plot(node_sizes, dfs_iter_matrix, marker='o', label='DFS Iterative (Matrix)')
plt.plot(node_sizes, dfs_rec_dict, marker='o', label='DFS Recursive (Dict)')
plt.plot(node_sizes, dfs_rec_matrix, marker='o', label='DFS Recursive (Matrix)')

# BFS data
plt.plot(node_sizes, bfs_iter_dict, marker='o', label='BFS Iterative (Dict)')
plt.plot(node_sizes, bfs_iter_matrix, marker='o', label='BFS Iterative (Matrix)')

# Adding title and labels
plt.title("DFS vs BFS Comparison")
plt.xlabel("Number of Nodes")
plt.ylabel("Time (seconds)")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend()
plt.show()


# Combine DFS and BFS for Dictionary and Matrix
# Create two lists: one for matrix times and one for dictionary times.
matrix_times = [sum(times) for times in zip(dfs_iter_matrix, dfs_rec_matrix, bfs_iter_matrix)]
dict_times = [sum(times) for times in zip(dfs_iter_dict, dfs_rec_dict, bfs_iter_dict)]

# Plot 2: Matrix vs Dictionary Combined (DFS + BFS for both)
plt.figure(figsize=(12, 6))

# Plot Matrix and Dictionary times
plt.plot(node_sizes, matrix_times, marker='o', label='Matrix (DFS + BFS)', linestyle='-', color='green')
plt.plot(node_sizes, dict_times, marker='o', label='Dictionary (DFS + BFS)', linestyle='-', color='blue')

# Adding title and labels
plt.title("Matrix vs Dictionary for DFS and BFS (Combined)")
plt.xlabel("Number of Nodes")
plt.ylabel("Time (seconds)")
plt.grid(True, linestyle="--", linewidth=0.5)
plt.legend()

plt.show()

```
